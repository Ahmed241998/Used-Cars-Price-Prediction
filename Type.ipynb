```python
# Python program to read
# json file
 
import json
 
# Opening JSON file
f = open('Car_Model_List.json')
 
# returns JSON object as 
# a dictionary
data = json.load(f)
```

```python
lst=[]
for i in data['results'] :
    lst.append([i['Make'],i['Category']])
```

```python
df=pd.DataFrame(data=lst,columns = ['Brand','Type'])
```

```python
dct={}
for i in df['Brand'].unique() :
    lst1=[]
    for j  in df[df['Brand'] == i]['Type'].tolist() :
        if ", " in j :
            lst1.extend(j.split(', '))
        else :
            if '/' in j :
                lst1.extend(j.split('/'))
            elif ',' in j :
                lst1.extend(j.split(','))
            else :
                lst1.append(j)
    dct[i.lower()] = lst1
```
